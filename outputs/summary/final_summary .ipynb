## 2007 年 YRBS 青少年睡眠時數與悲傷絕望感預測研究摘要
### 1.研究問題與動機（Research Question & Motivation）
本研究使用 2007 年美國青少年危險行為調查資料（YRBS 2007），探討高中生的每日睡眠時數是否能顯著預測其感到悲傷或絕望的可能性。由於悲傷或絕望感屬於二元結果變數，因此本研究採用 Logistic Regression 分析睡眠時數與心理健康風險之間的關係，並評估睡眠時數對悲傷或絕望感的預測效果。

### 2. 統計分析結果（Statistical Analysis Results）
#### 2.1 探索性資料分析（Exploratory Data Analysis）
經資料清洗與缺失值刪除後，本研究共保留：
有效分析樣本數（N） = 12,106
不同睡眠時數組別的悲傷或絕望感比例如下：
Sleep Group	Proportion of Sadness/Hopelessness
1	51.33%
2	41.05%
3	32.39%
4	25.35%
5	22.49%
6	25.54%
7	28.40%
整體而言，睡眠時間較短的學生具有較高的悲傷或絕望感比例，而隨著睡眠時間增加，該比例呈現下降趨勢。

#### 2.2 Logistic Regression Analysis
研究假設如下：
H₀：睡眠時數無法顯著預測高中生感到悲傷或絕望的可能性。
H₁：睡眠時數能夠顯著預測高中生感到悲傷或絕望的可能性。
本研究採用 Maximum Likelihood Estimation (MLE) 建立 Logistic Regression 模型。
主要結果如下：
Statistic	Value
Sample Size (N)	12,106
Sleep Coefficient (β)	-0.2475
Odds Ratio (OR)	0.7807
z-value	-16.366
p-value	< 0.001
Pseudo R²	0.0187

由於 p-value 小於 0.05，因此拒絕虛無假設（H₀），表示睡眠時數對高中生感到悲傷或絕望的可能性具有顯著預測效果。

### 3. 勝算比（Odds Ratio）解釋
睡眠時數的勝算比（Odds Ratio）為：
OR = 0.7807
此結果表示：
當睡眠時數組別每增加一個單位時，高中生感到悲傷或絕望的勝算平均下降約 21.9%。
計算方式如下：
(1 − 0.7807) × 100% = 21.93%

因此，較長的睡眠時間與較低的悲傷或絕望風險有關。

### 4. 結論（Conclusion）
本研究結果顯示，睡眠時數是高中生感到悲傷或絕望的重要預測因子。
Logistic Regression 分析指出，睡眠時數與悲傷或絕望感之間存在顯著負向關係（β = -0.2475, p < 0.001）。隨著睡眠時數增加，高中生感到悲傷或絕望的可能性顯著下降。
此外，勝算比（OR = 0.7807）顯示，每增加一個睡眠時數組別，感到悲傷或絕望的勝算平均下降約 21.9%。
整體而言，本研究支持睡眠時間較短的學生更容易出現悲傷或絕望感，而較充足的睡眠則與較低的情緒困擾風險有關。

## Executive Summary: Does Sleep Duration Predict Feelings of Sadness or Hopelessness Among High School Students? (YRBS 2007)
### 1. Research Question & Motivation
This study uses the 2007 Youth Risk Behavior Survey (YRBS) dataset to examine whether sleep duration significantly predicts the likelihood of feeling sad or hopeless among high school students. Because the outcome variable is binary, Logistic Regression was used to evaluate the relationship between sleep duration and emotional well-being.

### 2. Statistical Analysis Results
#### 2.1 Exploratory Data Analysis
After data cleaning and removing missing values, the final analytical sample contained:
Valid Sample Size (N) = 12,106
The proportions of sadness/hopelessness across sleep groups were:
Sleep Group	Proportion of Sadness/Hopelessness
1	51.33%
2	41.05%
3	32.39%
4	25.35%
5	22.49%
6	25.54%
7	28.40%
Overall, students with shorter sleep durations reported higher proportions of sadness or hopelessness.

#### 2.2 Logistic Regression Analysis
Statistical hypotheses:
H₀: Sleep duration does not significantly predict feelings of sadness or hopelessness.
H₁: Sleep duration significantly predicts feelings of sadness or hopelessness.
The Logistic Regression model was estimated using Maximum Likelihood Estimation (MLE).
Statistic	Value
Sample Size (N)	12,106
Sleep Coefficient (β)	-0.2475
Odds Ratio (OR)	0.7807
z-value	-16.366
p-value	< 0.001
Pseudo R²	0.0187
Since the p-value is less than 0.05, the null hypothesis was rejected.

### 3. Odds Ratio Interpretation
The odds ratio for sleep duration was:
OR = 0.7807
This indicates that for every one-category increase in sleep duration, the odds of feeling sad or hopeless decrease by approximately 21.9%.
Calculation:
(1 − 0.7807) × 100% = 21.93%
Therefore, longer sleep duration is associated with lower odds of emotional distress.

### 4. Conclusion
The findings indicate that sleep duration is a significant predictor of feelings of sadness or hopelessness among high school students.
The Logistic Regression model revealed a significant negative relationship between sleep duration and emotional distress (β = -0.2475, p < 0.001). As sleep duration increases, the likelihood of feeling sad or hopeless decreases.
Furthermore, the odds ratio (OR = 0.7807) suggests that each one-category increase in sleep duration is associated with an average 21.9% reduction in the odds of reporting sadness or hopelessness.
Overall, students with shorter sleep durations were more likely to experience sadness or hopelessness, whereas longer sleep durations were associated with lower emotional distress.
