<h1>Final Project</h1>
<h3>Research Question:Does sleep duration predict the likelihood of feeling sad or hopeless among high school students?</h3>

<h2>01 Data Preprocessing & Cleaning</h2>
<p><strong>1. Data Loading</strong></p>
<ul>
  <li>Used <code>pd.read_csv()</code> to load <code>YRBS_2007.csv</code>. The original sample size is <strong>14,041 records</strong>.</li>
</ul>

<p><strong>2. Variable Recoding</strong></p>
<ul>
  <li>The original dependent variable <code>SadOrHopeless</code> was coded as 1=Yes, 2=No.</li>
  <li>Created a new column <code>Sad_Recoded</code> to re-encode it into a dummy variable:
    <ul>
      <li><strong>1 = Sad/Hopeless</strong> (Originally 1.0)</li>
      <li><strong>0 = Not Sad/Hopeless</strong> (Originally 2.0, converted to 0.0)</li>
    </ul>
  </li>
</ul>

<p><strong>3. Data Cleaning & Missing Values</strong></p>
<ul>
  <li>Filtered <code>Sleep</code> and <code>Sad_Recoded</code>, and removed missing values using <code>.dropna()</code>.</li>
  <li>Retained a final valid sample size of <strong>12,106 records</strong> for analysis.</li>
</ul>

<p><strong>4. Data Export</strong></p>
<ul>
  <li>Exported the clean dataset to <code>data/cleaned_data.csv</code> using <code>.to_csv(..., index=False)</code> for future visualization and logistic regression analysis.</li>
</ul>

```python
import os
import pandas as pd
import numpy as np

# =====================================================================
# 1. 讀取原始資料 (Load Data)
# =====================================================================
try:
    df = pd.read_csv('YRBS_2007.csv')
except FileNotFoundError:
    df = pd.read_csv('data/YRBS_2007.csv')

print(f"🎉 成功讀取原始資料！總筆數：{len(df)} 筆")

# =====================================================================
# 2. 【核心步驟】資料重新編碼 (Data Recoding) ✨✨✨
# =====================================================================
# 根據 YRBS 說明書：SadOrHopeless 原始編碼為 1=Yes, 2=No
# 教授要求要重新編碼：我們要將 2(No) 轉換成 0，1(Yes) 維持 1，建立全新欄位 'Sad_Recoded'
recode_map = {1: 1, 2: 0}
df['Sad_Recoded'] = df['SadOrHopeless'].map(recode_map)

print("\n--- 🔍 重新編碼成果檢查 (Recoding Verification) ---")
print(df[['SadOrHopeless', 'Sad_Recoded']].head(10))
print("--------------------------------------------------\n")

# =====================================================================
# 3. 資料清理 - 處理遺失值 (Data Cleaning)
# =====================================================================
# 只選取自變數 'Sleep' 與我們剛剛重編碼好的應變數 'Sad_Recoded'，並刪除空值
cleaned_df = df[['Sleep', 'Sad_Recoded']].dropna()
print(f"✅ 資料清理完成！剔除空值後的最終有效分析資料筆數：{len(cleaned_df)} 筆")

# =====================================================================
# 4. 匯出乾淨資料檔案供 GitHub 呈現 (Export to CSV)
# =====================================================================
if not os.path.exists('data'):
    os.makedirs('data')

cleaned_df.to_csv('data/cleaned_data.csv', index=False)
print("💾 【第一步完全成功】包含『重新編碼』的乾淨資料已成功覆蓋匯出至 'data/cleaned_data.csv'！")
```
