## 變數定義與統計方法依據 (Variable Definitions & Statistical Methodology)

本文件詳細記錄本研究專案之變數定義、統計方法選擇依據，以及邏輯斯迴歸模型之各項統計假設檢驗。

---

### 1. 核心變數定義 (Core Variable Definitions)

#### 1.1 自變數 / 獨立變數 (Independent Variable)
* **原始問卷欄位**：`Q80`（資料集內命名為 `Sleep`）
* **變數描述**：受訪高中生在平常上學日的平均每日睡眠時數。
* **資料型態**：類別變數 / 序位變數 (Ordinal Variable)。
* **數值含意說明**：本變數為**「分類代碼」**（非直接小時數），各代碼對應之實際睡眠時間如下：
  * `1` = 4 小時或以下 (4 hours or less)
  * `2` = 5 小時 (5 hours)
  * `3` = 6 小時 (6 hours)
  * `4` = 7 小時 (7 hours)
  * `5` = 8 小時 (8 hours)
  * `6` = 9 小時 (9 hours)
  * `7` = 10 小時或以上 (10 or more hours)
* **分析處理**：保留原始 1–7 序位代碼，作為睡眠時數之序位尺度，以評估睡眠時數與悲傷或絕望感之間的關聯性。

#### 1.2 應變數 / 反應變數 (Dependent Variable)
* **原始問卷欄位**：`Q26`（資料集內重新編碼為 `Sad_Recoded`）
* **變數描述**：受訪高中生在過去 12 個月內，是否曾連續兩週以上幾乎每天感到悲傷或絕望。
* **資料型態**：二元變數 / 虛擬變數 (Binary / Dummy Variable)。
* **模型重編碼邏輯**：為符合邏輯斯迴歸之機率估計規範，數值定義如下：
  * `1` $\rightarrow$ **`1` (感到悲傷或絕望 / Yes)**
  * `2` $\rightarrow$ **`0` (未感到悲傷或絕望 / No)**

---

### 2. 統計方法選擇 (Statistical Method Selection)

* **選用方法**：二元邏輯斯迴歸分析 (Binary Logistic Regression)
* **選擇依據**：
  由於本研究的反應變數（`Sad_Recoded`）屬於**二元分類資料（非 0 即 1）**，傳統的普通最小平方法（OLS）線性迴歸會違反殘差常態性與變異數同質性假設，且預測機率可能超出 [0, 1] 的合理範圍。因此，本研究採用邏輯斯迴歸，透過 Logit 轉換將機率限縮於 0 至 1 之間，並以最大概似估計法（MLE）計算睡眠時數對悲傷機率的勝算比（Odds Ratio, OR），此為最符合本資料型態之標準推論統計方法。

---

### 3. 統計假設檢驗 (Statistical Assumptions Verification)

在進行推論統計前，本研究針對二元邏輯斯迴歸之各項基本假設評估如下：
1. **反應變數特徵**：應變數 `Sad_Recoded` 為嚴格的二元分佈（0 與 1），完全符合 Binary Logit 模型的設定需求。
2. **觀測值獨立性 (Independence)**：數據源自美國 CDC YRBSS 的分層隨機抽樣調查，各觀測值（個案學生）之間相互獨立，滿足獨立性假設。
3. **無過度共線性 (No Multicollinearity)**：本模型為單一自變數（`Sleep`）之簡單邏輯斯迴歸，不存在多重共線性問題。
4. **大樣本支持 (Large Sample Size)**：本模型排除缺失值後擁有 **12,106 筆**有效觀測值。極大的樣本量確保了最大概似估計（MLE）的漸近常態性，模型估計結果極為穩定且具備強大的統計檢定力（Statistical Power）。

## Variable Definitions & Statistical Methodology

This document explicitly logs the variable definitions, the rationale behind the statistical framework selection, and the verification of model assumptions for this research project.

---

### 1. Core Variable Definitions

#### 1.1 Independent Variable (X)
* **Original Survey Item**: `Q80` (Defined as `Sleep` in the dataset)
* **Description**: Average daily hours of sleep obtained by high school respondents on school nights.
* **Data Type**: Categorical / Ordinal Variable.
* **Value Interpretation**: This variable is structured as **categorical ordinal codes** (not direct numerical hours). The structural mapping is defined as follows:
  * `1` = 4 hours or less | `2` = 5 hours | `3` = 6 hours | `4` = 7 hours | `5` = 8 hours | `6` = 9 hours | `7` = 10 or more hours
* **Analytical Approach**:The original ordinal 1–7 coding was retained to examine the relationship between sleep duration and the likelihood of reporting sadness or hopelessness.

#### 1.2 Dependent Variable (Y)
* **Original Survey Item**: `Q26` (Recoded as `Sad_Recoded` in the dataset)
* **Description**: Whether the respondent felt persistently sad or hopeless for $\ge 2$ consecutive weeks within the past 12 months.
* **Data Type**: Binary / Dummy Variable (0 or 1).
* **Recoding Specifications**: 
  * Raw Code `1` (Yes) $\rightarrow$ **`1` (Experienced sadness/hopelessness)**
  * Raw Code `2` (No)  $\rightarrow$ **`0` (Did not experience sadness/hopelessness)**

---

### 2. Statistical Method Selection

* **Selected Framework**: Binary Logistic Regression
* **Methodological Rationale**:
  Because the dependent variable (`Sad_Recoded`) is strictly **binary (0 or 1)**, ordinary least squares (OLS) linear regression is mathematically inappropriate, as it violates homoscedasticity and normality of residuals, and can yield predicted probabilities outside the logical [0, 1] range. Therefore, Binary Logistic Regression is applied. Through the Logit link function, it models the bounded probability curve via Maximum Likelihood Estimation (MLE), providing mathematically sound coefficients and Odds Ratios (OR).

---

### 3. Statistical Assumptions Verification

Prior to conducting inferential modeling, the core assumptions of Binary Logistic Regression were verified:
1. **Binary Target Structure**: The dependent variable is strictly dichotomous (0/1), fully satisfying the framework requirements.
2. **Independence of Observations**: Data was collected via the CDC YRBSS multi-stage stratified random sampling design; all individual student respondents are mutually independent.
3. **Absence of Multicollinearity**: Since this is a simple logistic model with a single predictor (`Sleep`), multicollinearity is non-existent.
4. **Large Sample Properties**: With **12,106 valid observations** post-cleaning, the dataset easily satisfies the large-sample asymptotic properties required for stable MLE convergence, yielding exceptional statistical power.
